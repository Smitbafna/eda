```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np


sns.set(style='whitegrid')
plt.figure(figsize=(8, 5))
df = pd.read_csv('/kaggle/input/pcos-diagnosis-dataset/pcos_dataset.csv')
sns.scatterplot(x='Age', y='BMI', hue='PCOS_Diagnosis', data=df, palette='coolwarm')
plt.title('Age vs BMI (Colored by PCOS Diagnosis)')
plt.xlabel('Age')
plt.ylabel('BMI')
plt.legend(title='PCOS Diagnosis')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x='Testosterone_Level(ng/dL)', y='Antral_Follicle_Count', hue='PCOS_Diagnosis', data=df, palette='viridis')
plt.title('Testosterone Level vs Antral Follicle Count')
plt.xlabel('Testosterone Level (ng/dL)')
plt.ylabel('Antral Follicle Count')
plt.legend(title='PCOS Diagnosis')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.regplot(x='BMI', y='Testosterone_Level(ng/dL)', data=df, scatter_kws={'alpha':0.6}, line_kws={'color':'red'})
plt.title('BMI vs Testosterone Level')
plt.xlabel('BMI')
plt.ylabel('Testosterone Level (ng/dL)')
plt.show()
```

```python
pd.crosstab(df['Menstrual_Irregularity'], df['PCOS_Diagnosis']).plot(kind='bar', color=['#5c8eda','#ff6f61'])
plt.title('Menstrual Irregularity vs PCOS Diagnosis')
plt.xlabel('Menstrual Irregularity (1 = Irregular, 0 = Regular)')
plt.ylabel('Count')
plt.legend(['No PCOS', 'PCOS'])
plt.show()
```

```python
from scipy.stats import ttest_ind


pcos = df[df['PCOS_Diagnosis'] == 1]
non_pcos = df[df['PCOS_Diagnosis'] == 0]


continuous_vars = ['Age', 'BMI', 'Testosterone_Level(ng/dL)', 'Antral_Follicle_Count']


for var in continuous_vars:
    t_stat, p_value = ttest_ind(pcos[var], non_pcos[var])
    print(f'{var}: t-statistic = {t_stat:.2f}, p-value = {p_value:.4f}')
```

```python
from scipy.stats import chi2_contingency


contingency_table = pd.crosstab(df['Menstrual_Irregularity'], df['PCOS_Diagnosis'])


chi2, p, dof, expected = chi2_contingency(contingency_table)
print(f'Chi-squared Test:')
print(f'Chi2 value: {chi2:.2f}')
print(f'p-value: {p:.4f}')
```

```python
from scipy.stats import pearsonr


corr, p_value = pearsonr(df['BMI'], df['Testosterone_Level(ng/dL)'])
print(f'Pearson correlation (BMI vs Testosterone): {corr:.2f}, p-value: {p_value:.4f}')
```
